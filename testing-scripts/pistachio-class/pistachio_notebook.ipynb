We want to read in the pistachio data using csv and dataframes

```julia
data = CSV.read("../../testing-data/pistachio_class/pistachio.csv",DataFrame)
```

```julia
describe(data)
```

Need to get data in correct format

```julia
data_treet = data
data_treet.Class = Int32.(replace(data_treet.Class,"Siit_Pistachio"=>1,"Kirmizi_Pistachio"=>0))
```

```julia
data_select = select(data_treet,[:"ASPECT_RATIO",:"MINOR_AXIS",:"AREA",:"Class"])
data_select_norm = mapcols(zscore, data_select[!,Not("Class")])
data_select_norm[!,:Class] = data_select[!,:Class]
data_select_norm
```

```julia
train,test = partition(data_select_norm, 0.8, shuffle=true);
```

```julia
train_in = Matrix(train[!,Not("Class")])
n_points,dim = size(train_in)
train_target = Array(train[!,"Class"])
train_target = hcat(train_target,zeros(length(train_target),dim-1))

test_in = Matrix(test[!,Not("Class")])
test_target = Array(test[!,"Class"])
test_target = hcat(test_target,zeros(length(test_target),dim-1))

print(size(train_in))
print(size(train_target))

print(size(test_in))
print(size(test_target))
```

```julia
include("TransformM.jl")
using .TransformM
```

```julia
tf = TransformM.Transform_Const(dim,train_in,train_target)
```

```julia
pred_values = zeros(0)
preds = zeros(0)
npreds = size(test_target,1)
npreds = 200
#for i in ProgressBar(1:npreds)
for i in 1:npreds
    test_i_val = test_in[i,:]
    pred_i_array = TransformM.map_point(tf,test_in[i,:])
    pred_i_val = pred_i_array[1]

    pred_i=0
    if pred_i_val>0
        pred_i=1
    end

    append!(pred_values,pred_i_val)
    append!(preds,pred_i)

    #print("Pred: $i/$npreds \r")
    #flush(stdout)
    #print("Pred: $i/$npreds")
    #print("\e[2K") # clear whole line
    #print("\e[1G") # move cursor to column 1
end
frac_correct = sum(abs.(-preds.-test_target[1:npreds,1].+1))/npreds

```
